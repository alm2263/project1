## database schema

```mysql
create table businesses(
  business_id varchar(22) primary key,
  business_name text,
  neighborhood text,
  city text,
  state text,
  address text,
  postal_code varchar(7),
  latitude float,
  longitude float,
  stars numeric,
  review_count int,
  is_open int,
  business_parking_garage boolean,
  business_parking_street boolean,
  business_parking_validated boolean,
  business_parking_lot boolean,
  business_parking_valet boolean,
  restaurants_price_range2 text,
  noise_level text,
  restaurants_attire text,
  monday_hour text,
  tuesday_hour text,
  wednesday_hour text,
  thursday_hour text,
  friday_hour text,
  saturday_hour text,
  sunday_hour text
);


create table business_category(
  business_id varchar(22),
  category_name text,
  primary key (business_id, category_name)
);


create table photos(
  photo_id varchar(22),
  business_id varchar (22) references Businesses,
  label text,
  primary key (photo_id, business_id) 
);

create table reviews(
  review_id char(22),
  business_id char(22),
  review_date date,
  stars int,
  cool int,
  funny int,
  review_text text
);

create table users( 
  user_id char(22) primary key, 
  name text ,
  review_count int, 
  yelping_since date, 
  useful int(11) ,
  funny int(11) ,
  cool int(11) ,
  fans int(11) ,
  average_stars float ,
  compliment_more int(11) ,
  compliment_profile int(11) ,
  compliment_cut int(11), 
  compliment_list int(11), 
  compliment_note int(11) ,
  compliment_plain int(11) ,
  compliment_cool int(11) ,
  compliment_funny int(11) ,
  compliment_writer int(11))
```
